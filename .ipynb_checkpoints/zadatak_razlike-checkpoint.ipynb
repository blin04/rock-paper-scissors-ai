```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

<h2>Uƒçitati slike i prikazati ih</h2>

```python
r1 = cv2.imread('slike/razlike_1.png')
r1 = cv2.cvtColor(r1, cv2.COLOR_BGR2RGB)

r2 = cv2.imread('slike/razlike_2.png')
r2 = cv2.cvtColor(r2, cv2.COLOR_BGR2RGB)

fig , axarr = plt.subplots(1,2)
fig.set_figheight(15)
fig.set_figwidth(15)
axarr[0].imshow(r1)
axarr[1].imshow(r2)
```

<h2>Uzeti samo jedan kanal slike</h2>

```python
r1_0 = r1[:,:,0]
r2_0 = r2[:,:,0]
```

<h2>Oduzeti slike (naƒái razliku matrica)</h2>

```python
res = r2_0-r1_0
```

<h2>Podebljati dobijene regione</h2>

```python
kernel = np.ones((5,5), np.uint8)
res1 = cv2.dilate(res, kernel, iterations = 1)

fig , axarr = plt.subplots(1,2)
fig.set_figheight(15)
fig.set_figwidth(15)
axarr[0].imshow(res)
axarr[1].imshow(res1)
```

<h2>Pronaƒái konture</h2>

```python
contours, r = cv2.findContours(res1, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)
```

<h2>Nacrtati krugove na originalnoj slici</h2>

```python
# Inicijalizacija pomoƒánih promenljivih
contours_poly = [None]*len(contours)
centers = [None]*len(contours)
radius = [None]*len(contours)

# Iteriramo kroz konture
for i, c in enumerate(contours):
    # aproksimiramo konturu poligonom
    contours_poly[i] = cv2.approxPolyDP(c, 3, True)
    # tra≈æimo najmanji krug koji obuhvata pronaƒëenu konturu
    centers[i], radius[i] = cv2.minEnclosingCircle(contours_poly[i])
    
# Crtamo krugove
for c,r in zip(centers, radius):   
    cv2.circle(r1, (int(c[0]), int(c[1])), int(r), (255,0,0), 2)
    
    
plt.figure(figsize=(15, 6))
plt.imshow(r1)
```
