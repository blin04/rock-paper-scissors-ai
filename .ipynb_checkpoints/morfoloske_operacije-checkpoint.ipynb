```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt
```

<h2>Erozija</h2>

```python
slika = cv2.imread('slike/j.png',0)
kernel = np.ones((5,5), np.uint8)
erodirana_slika = cv2.erode(slika, kernel, iterations = 1)

fig , axarr = plt.subplots(1,2)
fig.set_figheight(15)
fig.set_figwidth(15)
axarr[0].imshow(slika, cmap='gray')
axarr[1].imshow(erodirana_slika, cmap='gray')
```

<h2>Dilatacija</h2>

```python
slika = cv2.imread('slike/j.png',0)
kernel = np.ones((5,5), np.uint8)
dilatacija = cv2.dilate(slika, kernel, iterations = 1)

fig , axarr = plt.subplots(1,2)
fig.set_figheight(15)
fig.set_figwidth(15)
axarr[0].imshow(slika, cmap='gray')
axarr[1].imshow(dilatacija, cmap='gray')
```
