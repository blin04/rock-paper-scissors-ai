```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

<h2>Uƒçitavamo sliku</h2>

```python
slika = cv2.imread('slike/pare.jpg')
slika = cv2.cvtColor(slika, cv2.COLOR_BGR2RGB)

slika_s = slika.copy()

# Malo blurujemo kako bismo smanjili sum
slika = cv2.blur(slika, (3, 3))

plt.figure(figsize=(25, 10))
plt.imshow(slika)
```

```python
s = slika[:,:,0]
#ret, s = cv2.threshold(s, 130, 255, cv2.THRESH_BINARY_INV)
#plt.imshow(s)
```

<h2>Tra≈æimo krugove</h2>

```python
circles = cv2.HoughCircles(s, dp=1, minRadius=30, maxRadius=90, minDist=30, param1=20, param2=85, method = cv2.HOUGH_GRADIENT)

if circles is not None:
    for circle in circles[0]:
        x, y, r = list(circle)
        # draw the outer circle
        cv2.circle(slika,(x,y), r,(255, 0, 0), 2)
        # draw the center of the circle
```

```python
plt.figure(figsize=(25, 10))
plt.imshow(slika)
```

```python
# izvlacimo r za svaku kovanicu kako bismo ih sortirali i definisali opsege
radius_lists = []
if circles is not None:
    for circle in circles[0]:
        x, y, r = list(circle)
        radius_lists.append(round(r))
        
radius_lists.sort()
print(radius_lists)
```

```python
def vrednost_kovanice(r):
    if r >= 61:
        return 20, (255, 0, 0)
    elif r >= 55 and r < 61:
        return 5, (0, 255, 0)
    elif r >= 52 and r < 55:
        return 2, (0, 0, 255)
    elif r < 52:
        return 1, (255, 255, 255)
    else:
        return 0
```

<h2>Prikazujemo vrednost kovanice i krug</h2>

```python
if circles is not None:
    for circle in circles[0]:
        x, y, r = list(circle)
        v, c = vrednost_kovanice(r)
        cv2.circle(slika_s,(x,y), r, c, 2)
        cv2.putText(slika_s, str(v), (x,y), cv2.FONT_HERSHEY_SIMPLEX, 2, c, 2, cv2.LINE_AA)

plt.figure(figsize=(25, 10))
plt.imshow(slika_s)
```
