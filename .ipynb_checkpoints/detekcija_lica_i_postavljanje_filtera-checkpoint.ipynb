```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

<h2>U훾itavamo modele</h2>

```python
face_cascade = cv2.CascadeClassifier('modeli/haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier('modeli/haarcascade_eye.xml')
smile_cascade = cv2.CascadeClassifier('modeli/haarcascade_smile.xml')
```

<h2>U훾itavamo slike</h2>

```python
oko = cv2.imread('slike/eye.png')
# Menjamo veli훾inu sli훾ice oka
oko = cv2.resize(oko, None, fx=0.3, fy=0.3)
oko_visina, oko_sirina, oko_kanali = oko.shape

plt.imshow(cv2.cvtColor(oko, cv2.COLOR_BGR2RGB))
print(oko.shape)
```

```python
slika = cv2.imread('slike/jokic.png')
plt.imshow(cv2.cvtColor(slika, cv2.COLOR_BGR2RGB))
```

```python
# Konvertujemo sliku u crno belu
gray = cv2.cvtColor(slika, cv2.COLOR_BGR2GRAY)
```

<h2>Tra탑imo lica</h2>

```python
# Tra탑imo lica
lica = face_cascade.detectMultiScale(gray, 1.3, 5)
for (x, y, w, h) in lica:
    roi_gray = gray[y:y + h, x:x + w]
    roi_color = slika[y:y + h, x:x + w]
    
    # Tra탑imo o훾i
    oci = eye_cascade.detectMultiScale(roi_gray)
    for (x_oko, y_oko, w_oko, h_oko) in oci:
        t = y_oko + y - 15
        l = x_oko + x -15

        for i in range(3):
            roi = slika[t:t + oko_visina, l:l + oko_sirina, i]
            sivo_oko = cv2.cvtColor(oko, cv2.COLOR_BGR2GRAY)
            ret, maska_oka = cv2.threshold(sivo_oko, 3, 255, cv2.THRESH_BINARY)
            maska_inv = cv2.bitwise_not(maska_oka)

            img1_bg = cv2.bitwise_and(roi, roi, mask=maska_inv)
            img2_fg = cv2.bitwise_and(oko, oko, mask=maska_oka)

            lice_sa_okom = cv2.add(img1_bg, img2_fg[...,i])
            slika[t:t + oko_visina, l:l + oko_sirina, i] = lice_sa_okom
```

```python
plt.imshow(cv2.cvtColor(slika, cv2.COLOR_BGR2RGB))
```
