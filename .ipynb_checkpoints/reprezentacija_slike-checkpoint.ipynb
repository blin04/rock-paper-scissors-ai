<h2>Učitavamo još jednu sliku</h2>

```python
jedrilica = cv2.imread('slike/jedrilica.png')
jedrilica = cv2.cvtColor(jedrilica, cv2.COLOR_BGR2RGB)

plt.imshow(jedrilica)
```

<h2>Zbir slika</h2>

```python
plt.imshow(more+jedrilica)
```

<h2>Maskiranje</h2>

```python
ret, maska = cv2.threshold(jedrilica, 0, 255, cv2.THRESH_BINARY_INV)
# Napomena: maska mora da bude dvodimenzioni niz tipa uint8
# uzimamo nulti kanal maske
maska = maska[:,:,0]
# konvertujemo tip u uint8
maska = maska.astype('uint8')
plt.title("Maska jedrilice")
plt.imshow(maska)
```

```python
# maskirano = mora - maska
maskirano = cv2.bitwise_or(more, more, mask=maska)
plt.title("More sa rupom")
plt.imshow(maskirano)
```

```python
plt.title("Jedrilica na moru")
plt.imshow(maskirano+jedrilica) 
```

<h2>Bojenje delova slike</h2>

```python
jedrilica_na_moru = maskirano+jedrilica
jedrilica_na_moru[30:80, 20:100,:] = [255,0,0]
plt.imshow(jedrilica_na_moru) 
```

<h2>Zadatak 1 - Razlike</h2>

```python
# Ucitati slike
r1 = cv2.imread('slike/razlike_1.png')
r1 = cv2.cvtColor(r1, cv2.COLOR_BGR2RGB)

r2 = cv2.imread('slike/razlike_2.png')
r2 = cv2.cvtColor(r2, cv2.COLOR_BGR2RGB)

fig , axarr = plt.subplots(1,2)
fig.set_figheight(15)
fig.set_figwidth(15)
axarr[0].imshow(r1)
axarr[1].imshow(r2)
```

```python
r1_0 = r1[:,:,0]
r2_0 = r2[:,:,0]
```

```python
# pronaci razlike
res = r2_0-r1_0

kernel = np.ones((5,5), np.uint8)
res1 = cv2.dilate(res, kernel, iterations = 1)

fig , axarr = plt.subplots(1,2)
fig.set_figheight(15)
fig.set_figwidth(15)
axarr[0].imshow(res)
axarr[1].imshow(res1)
```

```python
# pronaci konture
contours, r = cv2.findContours(res1, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)
```

```python
# plotovati krugove na originalnoj slici
contours_poly = [None]*len(contours)
centers = [None]*len(contours)
radius = [None]*len(contours)
for i, c in enumerate(contours):
    contours_poly[i] = cv2.approxPolyDP(c, 3, True)
    centers[i], radius[i] = cv2.minEnclosingCircle(contours_poly[i])
    

for c,r in zip(centers, radius):   
    cv2.circle(r1, (int(c[0]), int(c[1])), int(r), (255,0,0), 2)
    
    
plt.figure(figsize=(15, 6))
plt.imshow(r1)
```
