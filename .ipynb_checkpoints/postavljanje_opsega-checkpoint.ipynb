```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
slika = cv2.imread('slike/luna_rosa.jpeg')
crno_bela_slika = cv2.cvtColor(slika, cv2.COLOR_BGR2GRAY)

ret, opseg = cv2.threshold(crno_bela_slika, 150, 255, cv2.THRESH_BINARY)


fig , axarr = plt.subplots(1,2)
fig.set_figheight(15)
fig.set_figwidth(15)
axarr[0].imshow(crno_bela_slika, cmap='gray')
axarr[1].imshow(opseg, cmap='gray')
```

```python
slika = cv2.imread('slike/luna_rosa.jpeg')
crno_bela_slika = cv2.cvtColor(slika, cv2.COLOR_BGR2GRAY)

ret, opseg = cv2.threshold(crno_bela_slika, 150, 255, cv2.THRESH_BINARY_INV)


fig , axarr = plt.subplots(1,2)
fig.set_figheight(15)
fig.set_figwidth(15)
axarr[0].imshow(crno_bela_slika, cmap='gray')
axarr[1].imshow(opseg, cmap='gray')
```

```python
slika = cv2.imread('slike/luna_rosa.jpeg')
crno_bela_slika = cv2.cvtColor(slika, cv2.COLOR_BGR2GRAY)


th2 = cv2.adaptiveThreshold(crno_bela_slika, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY,11,2)

fig , axarr = plt.subplots(1,2)
fig.set_figheight(15)
fig.set_figwidth(15)
axarr[0].imshow(crno_bela_slika, cmap='gray')
axarr[1].imshow(th2, cmap='gray')
```

```python
slika = cv2.imread('slike/luna_rosa.jpeg')
crno_bela_slika = cv2.cvtColor(slika, cv2.COLOR_BGR2GRAY)

th3 = cv2.adaptiveThreshold(crno_bela_slika, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY,11,2)

fig , axarr = plt.subplots(1,2)
fig.set_figheight(15)
fig.set_figwidth(15)
axarr[0].imshow(crno_bela_slika, cmap='gray')
axarr[1].imshow(th3, cmap='gray')

```
