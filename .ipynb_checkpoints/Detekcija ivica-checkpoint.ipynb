```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
slika = cv2.imread('slike/ruka.jpg',0)
#slika = cv2.cvtColor(slika, cv2.COLOR_BGR2RGB)
ivice = cv2.Canny(slika, 150, 300)

fig , axarr = plt.subplots(1,2)
fig.set_figheight(15)
fig.set_figwidth(15)
axarr[0].imshow(slika)
axarr[1].imshow(ivice)
```
