```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

# OpenCV konture
https://docs.opencv.org/3.4/d4/d73/tutorial_py_contours_begin.html

```python
# ucitavamo sliku
slika = cv2.imread('slike/ruka.jpg')
kopija_slike = slika.copy()
cbs = cv2.cvtColor(slika, cv2.COLOR_BGR2GRAY)

# plot
plt.imshow(cbs, cmap='gray')
```

```python
# Postavljamo opseg kako bi lakse pronasli konture
ret, so = cv2.threshold(cbs, 230, 255, cv2.THRESH_BINARY)
plt.imshow(so, cmap = 'gray')
```

```python
# Trazimo konture
konture, h = cv2.findContours(so, cv2.RETR_TREE, cv2.CHAIN_APPROX_NONE)
# Crtamo konture na slici
cv2.drawContours(kopija_slike, konture, -1, (255, 0, 0),3)

#plot slika
fig , axarr = plt.subplots(1,2)
fig.set_figheight(15)
fig.set_figwidth(15)
axarr[0].imshow(slika)
axarr[1].imshow(kopija_slike)
```

```python
# Trazimo konture
konture, h = cv2.findContours(so, cv2.RETR_TREE, cv2.CHAIN_APPROX_NONE)
# Crtamo konture na slici
cv2.drawContours(kopija_slike, konture, -1, (0, 0, 255),3)

#plot slika
fig , axarr = plt.subplots(1,2)
fig.set_figheight(15)
fig.set_figwidth(15)
axarr[0].imshow(cv2.cvtColor(slika, cv2.COLOR_BGR2RGB))
axarr[1].imshow(cv2.cvtColor(kopija_slike, cv2.COLOR_BGR2RGB))
```

<h2>Crtamo samo konturu na beloj slici</h2>

```python
h, w, c = slika.shape
prazna_slika = np.zeros([w,h,c],dtype=np.uint8)
cv2.drawContours(prazna_slika, konture, -1, (255, 255, 255),3)

plt.imshow(prazna_slika)
```
